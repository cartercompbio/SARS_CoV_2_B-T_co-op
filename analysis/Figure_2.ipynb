```python
def get_position_affinity_score(frac_df, affinity_col='min_median_affinity', affinity_col_prefix='Median affinity'):
    pos_min_median_list = []
    
    # add pos data to aff data
    frac_df['pos_range'] = frac_df.apply(lambda x: range(x['pos'],x['pos']+len(x['Peptide'])), axis=1)

    # get position affinity scores
    for position in range(len(sp)):
        # subset to any peptides covering position
        pos_df = frac_df[frac_df['pos_range'].apply(lambda x: position in x)]
        pos_df = pos_df.set_index(['Peptide','pos_range'])

        # append
        median_col = [x for x in frac_df.columns if affinity_col_prefix in x][0]
        pos_min_median_list.append(pos_df[median_col].min())

    # aggregated df
    position_aff_df = pd.DataFrame({'position': range(len(sp)), 
                                    affinity_col: pos_min_median_list})
    return position_aff_df
```

##### Additionally filter for closeness to FNCY +/-50aa

```python
def filter_for_close_to_FNCY(frac_df, melted_df, plusminus=50, binding_thresh=10):
    num_alleles = len(melted_df['Allele'].unique())
    print('{} alleles'.format(num_alleles))
    
    frac_df['end_pos'] = frac_df.apply(lambda x: x['pos']+x['len'], axis=1)
    frac_df['center_pos'] = frac_df.apply(lambda x: int(np.median([x['pos'],x['end_pos']])), axis=1)
    frac_df['distance_to_FNCY'] = frac_df['center_pos'].apply(lambda x: abs(x-sp.find('FNCY')+1))
    
    close_peptides = frac_df[frac_df['distance_to_FNCY']<=plusminus]['Peptide'].values

    close_melted_df = melted_df[melted_df['Peptide'].isin(close_peptides)]
    close_frac_df = frac_df[frac_df['Peptide'].isin(close_peptides)]

    return close_frac_df, close_melted_df
```

```python
close_frac_df, close_aff_df = filter_for_close_to_FNCY(frac_df, aff_df)
print(close_frac_df.shape, close_aff_df.shape)

close_frac_common_df, close_aff_common_df = filter_for_close_to_FNCY(frac_common_df, aff_common_df)
print(close_frac_common_df.shape, close_aff_common_df.shape)
```

##### Add color col

```python
color_df = pd.DataFrame.from_dict(region_domains,orient='index')
color_df.columns = ['start_pos', 'end_pos', 'color']
color_df['position'] = color_df.apply(lambda x: list(range(x['start_pos'],x['end_pos'])), axis=1)
color_df = color_df.explode('position')

priority_color_order = ['RBD motif (437-508)', 'RBD (319-541)', 'Fusion peptide 1 (816-837)', 
                       'Fusion peptide 2 (835-855)', 'Heptad repeat 1 (920-970)', 'Heptad repeat 2 (1163-1202)']
color_df = color_df.loc[priority_color_order]
color_df = color_df.drop_duplicates(subset=['position']) # if multiple domains cover position, keep RBD first

color_df.head()
```

```python
def add_color_col(df, color_df):
    pos_col = [x for x in df.columns if x.startswith('pos')][0]
    df = pd.merge(df, color_df, left_on=pos_col, right_on='position', how='left')
    df['color'] = df['color'].fillna('white')
    return df
```

```python
close_frac_df = add_color_col(close_frac_df, color_df)
close_frac_common_df = add_color_col(close_frac_common_df, color_df)


# close_aff_df = add_color_col(close_aff_df, color_df)
# close_aff_common_df = add_color_col(close_aff_common_df, color_df)

frac_common_df = add_color_col(frac_common_df, color_df)
frac_df = add_color_col(frac_df, color_df)


# aff_common_df = add_color_col(aff_common_df, color_df)
```

```python
def plot_frac_presentable_top_N(frac_df, melted_df, N, f, shorter=True):
    '''Lower panels (barplot, boxplot)'''
    inner_grid = gridspec.GridSpecFromSubplotSpec(2,1, f, height_ratios=[1,3.3])

    a1 = plt.subplot(inner_grid[0,0])
    a2 = plt.subplot(inner_grid[1,0])

    # fraction
    frac_df = frac_df.sort_values(by='Fraction of binding alleles', ascending=False)
    barp = sns.barplot(x='Peptide', y='Fraction of binding alleles', data=frac_df.iloc[:N], ax=a1, )
#     barp = a1.bar(frac_df.iloc[:N]['Peptide'].values, 
#                    frac_df.iloc[:N]['Fraction of binding alleles'].values, 
#                    color=frac_df.iloc[:N]['color'].values, edgecolor='k', alpha=0.25)
    a1.set_xticklabels([]*N, rotation=45, ha='right'), 
    a1.set_xlabel('')
    a1.set_ylim(0,1)
    
    # set color
    for i, bar in enumerate(barp.patches):
        bar.set_color(frac_df.iloc[:N]['color'].values[i])
        bar.set_edgecolor('darkgrey')
        bar.set_alpha(0.25)


    # top 30 peptides - are they super well presented?
    top_N_peptides = frac_df.iloc[:N]['Peptide'].unique()
    top_N_df = melted_df[melted_df['Peptide'].isin(top_N_peptides)]
    a1.set_ylabel('Fraction of\nbinding alleles')

    # label median score
    fracs = frac_df.iloc[:30]['Fraction of binding alleles'].apply(lambda x: \
                                                                   '{}'.format(int(round(x*100,0)))).values
    vertical_offset = 0.5 # offset from median for display
    if shorter:
        vertical_offset = 0.1

    for xtick in a1.get_xticks():
        a1.text(xtick, vertical_offset, fracs[xtick], 
                horizontalalignment='center',size='xx-small',color='k',)



    ### bottom

    bp = sns.boxplot(x='Peptide', y='Affinity', data=top_N_df, order=top_N_peptides, ax=a2,
                     medianprops={'color':'darkgrey', 'alpha': 0.25},
                     flierprops=dict(markerfacecolor='darkgrey', markeredgecolor='darkgrey'))
    # set color
    for i in range(N):
        mybox = bp.artists[i]
        mybox.set_facecolor(frac_df.iloc[:N]['color'].values[i])
        mybox.set_edgecolor('darkgrey')
        mybox.set_alpha(0.25)
        
    
#     a2.set_xticklabels(a2.xticks(), rotation=45, ha='right'), 
    plt.xticks(rotation=45, ha='right')
    a2.set_xlabel('Top 30 peptides')
    a2.set_ylabel('Percentile\naffinity score')
    a2.set_ylim(-1,52) 
    a2.axhline(10, color='grey', ls='--', lw=1.1)

    # label median score
    medians = top_N_df.groupby(['Peptide'])['Affinity'].median().apply(lambda x: int(round(x,0)))
    medians = medians[top_N_peptides] # order
    vertical_offset = top_N_df['Affinity'].median() * 0.02 # offset from median for display


    for xtick in bp.get_xticks():
        bp.text(xtick, medians[xtick] + vertical_offset, medians[xtick], 
                horizontalalignment='center',size='xx-small',color='k',)
    
```

```python
fig = plt.figure(constrained_layout=True, figsize=(12,15))
gs = gridspec.GridSpec(15,6, figure=fig)

#################################################
# Panel B: all vs common allele median affinity #
#################################################
f3_ax0 = fig.add_subplot(gs[0:2,4:6])
scp = sns.scatterplot(x=pos_common_df['min_median_affinity'].values, 
                      y=pos_df['min_median_affinity'].values, ax=f3_ax0, palette='Oranges')
r, p = pearsonr(x=pos_common_df['min_median_affinity'].values, y=pos_df['min_median_affinity'].values)
f3_ax0.text(5, 85, 'r={:.2}, p={:.2}'.format(r,p))
f3_ax0.set_xlim(0,100), f3_ax0.set_ylim(0,100)

################################
# Panel C: positional affinity #
################################

f3_ax1 = fig.add_subplot(gs[2:6, 0:5])
f3_ax1.plot(pos_common_df['position'].values, pos_common_df['min_median_affinity'].values, 
            label='Position affinity score', color='purple')
f3_ax1.axhline(10, ls='--', color='grey', lw=1.1)

# glycan background
for i, region in enumerate(glycan_regions):
    if i == 0:
        f3_ax1.axvspan(region[0], region[-1], facecolor='darkgrey', alpha=0.1, label='Glycans')
    else:
        f3_ax1.axvspan(region[0], region[-1], facecolor='darkgrey', alpha=0.1)

# domain regions background
for region in region_domains:
    f3_ax1.axvspan(region_domains[region][0], region_domains[region][1], alpha=0.2, 
                label=region, color=region_domains[region][2])

# params
f3_ax1.set_xlim(0,len(sp)), f3_ax1.set_ylim(0,100)
f3_ax1.set_xlabel('SARS CoV-2 spike protein positions'), f3_ax1.set_ylabel('Best peptide spanning position')
f3_ax1.legend(bbox_to_anchor=(1,0.9), prop={'size': 8})

    
####################
# Panel D: glycans #
####################

f3_ax2 = fig.add_subplot(gs[6, 0:5])
sns.heatmap(glycan_heatmap_data, cbar=False, cbar_kws={"shrink": .7}, yticklabels=True, cmap='Oranges', 
            ax=f3_ax2)

# plot params    
plt.tight_layout()

savepath = '../data/figures/Fig3B-D.pdf'
print('Saving to {}'.format(savepath))
plt.savefig(savepath, bbox_inches='tight')
plt.show()
```
